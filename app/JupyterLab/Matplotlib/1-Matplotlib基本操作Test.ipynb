```python
x=np.linspace(-10,10)
y=np.sin(x)
plt.plot(x,y,linewidth = 1.0)
plt.show()
```

```python
plt.plot(x,y,color='b',linestyle=':',marker='o',markerfacecolor='r',markersize=10)
plt.show()
```

```python
line=plt.plot(x,y)
plt.setp(line,color='r',linewidth=2.0,alpha=0.4)
plt.show()
```
